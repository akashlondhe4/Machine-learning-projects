```python
import pandas as pd
import numpy as np

# import plotting libraries
import matplotlib
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
%matplotlib inline  
```

```python
location = r"F:\\AKASHRAJ\\AKASH STUDY MATERIAL\\Project file\\311_Service.csv"
data = pd.read_csv(location)
data.head()
```

```python
data.shape
```

```python
fig,ax=plt.subplots(figsize=(12,7))
data.groupby(['Complaint Type']).count()['Unique Key'].plot.bar(ax=ax, width=1.5)
```

```python
# plot the bar graph based on complaints

fig,ax=plt.subplots(figsize=(25,15))
newdata_C=pd.DataFrame(data.groupby(['Complaint Type']).count())
newdata_C['Unique Key'].plot.bar(ax=ax,width=1.5)
```

```python
#Find out most occuring complaints.

newdata_C.loc[newdata_C['Unique Key']==max(newdata_C['Unique Key']),'Unique Key']
```
